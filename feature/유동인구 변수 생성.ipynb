```python
tmp3
```

```python
tmp3['resd_pop'] = tmp3['resd_pop_cnt_x'] + tmp3['resd_pop_cnt_y']
tmp3['work_pop'] = tmp3['work_pop_cnt_x'] + tmp3['work_pop_cnt_y']
```

```python
tmp3.drop(['resd_pop_cnt_x','resd_pop_cnt_y','work_pop_cnt_x','work_pop_cnt_y'],axis=1,inplace=True)

tmp3['visit_pop'] = tmp3['visit_pop'] + tmp3['visit_pop_cnt']

tmp3.drop(['visit_pop_cnt'],axis=1,inplace=True)
tmp3
```

```python
# emd_nm == '전체' 삽입
tmp_all = tmp3.groupby(['year','month','date']).sum().reset_index()
tmp_all['emd_nm'] = "전체"
```

```python
tmp_all
```

```python
group_df = pd.merge(tmp3,tmp_all,on=('year','month','date','emd_nm'),how='outer')

group_df
```

```python
group_df[group_df.visit_pop_x.isna()].emd_nm.unique()
```

```python
group_df = group_df.fillna(0)
group_df['visit_pop'] = group_df['visit_pop_x'] + group_df['visit_pop_y']
group_df['resd_pop'] = group_df['resd_pop_x'] + group_df['resd_pop_y']
group_df['work_pop'] = group_df['work_pop_x'] + group_df['work_pop_y']
```

```python
group_df = group_df.rename(columns={'em_cnt_x':'em_cnt','em_g_x':'em_g','pay_amt_x':'pay_amt'})
```

```python
group_df.drop(['em_cnt_y','em_g_y','pay_amt_y','visit_pop_y','resd_pop_y','work_pop_y','visit_pop_x','resd_pop_x','work_pop_x'],axis=1,inplace=True)
```
